```python
import requests
from requests.auth import HTTPBasicAuth
import pandas as pd
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
REED_USER = os.getenv('REED_USER')
REED_PASSWORD = os.getenv('REED_PASSWORD')
REED_BASE_URL = "https://www.reed.co.uk/api/1.0/search/"
```

```python
params = {
    "keywords" : "data science",
}
auth = HTTPBasicAuth(REED_USER, REED_PASSWORD)
response = requests.get(REED_BASE_URL, auth=auth, params=params)
print(response.status_code)
response_data = response.json()
print(response_data.keys())
```

```python
jobs_df = pd.json_normalize(response_data['results'])
jobs_df.head()
```

```python
df_selected = jobs_df[['jobId', 'employerId','employerName', 'jobTitle', 'locationName', 'minimumSalary', 'maximumSalary', 'currency', 'date', 'applications', 'jobUrl']].copy()
df_selected['date'] = pd.to_datetime(df_selected['date'], format="%d/%m/%Y", errors='coerce')
df_selected.head()
```
