```python
from pathlib import Path
import pandas as pd
```

```python
# TODO download from https://www.doogal.co.uk/files/postcodes.zip
# !!! the raw file is huge, it's about 1.9GB
POSTCODES_FOLDER = Path("../../temp/postcodes")
POSTCODES_PATH = POSTCODES_FOLDER / Path("postcodes.csv")
```

```python
# Load first 1000 postcodes
postcodes_df = (pd.read_csv(POSTCODES_PATH)[lambda df: df["In Use?"].eq("Yes")])

print(postcodes_df.columns )
postcodes_df.head()
```

```python
postcodes_df.to_csv(POSTCODES_FOLDER / Path("postcodes_inuse.csv"), index=False)
```

```python
# Get human readable data
postcodes_filtered_df = postcodes_df[['Postcode', 'Built up area',
       'County', 'District', 'Ward',
       'Country', 'Constituency']].copy()
# remove spaces from Postcodes
postcodes_filtered_df['Postcode'] = postcodes_filtered_df['Postcode'].str.replace(' ', '')
postcodes_filtered_df.head()
```

```python
postcodes_filtered_df.to_csv(POSTCODES_FOLDER / Path("postcodes_inuse_filtered.csv"), index=False)
```
